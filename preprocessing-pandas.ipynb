**Dealing Missing Values**

To clean data we will drop the rows and columns that are null. we will dropna function to drop rows or columns to clean data. 

Dropping Rows

```python
import pandas as pd

df = pd.read_csv('laptopData.csv')

print("\nbefore")
print(df.shape) 
print(df.isnull().sum())

```

Dropping Columns

```python
#using dropna function
df.dropna(axis='index', inplace=True)       
print("after")
print(df.shape) 
print(df.isnull().sum())
```

```python
import pandas as pd

df = pd.read_csv('laptopData.csv')
print("before")
print(df.shape)
print(df.isnull().sum())
```

```python
df.dropna(axis='columns', inplace=True)

print("after")
print(df.shape)
print(df.isnull().sum())
```

**Duplicate Values**

```python
import pandas as pd

df = pd.read_csv('laptopData.csv')

print("Before dropping duplicates")
df
```

```python
print("After dropping duplicates")
df.drop_duplicates()
```

The duplicate values are checked and dropped by the above function.
